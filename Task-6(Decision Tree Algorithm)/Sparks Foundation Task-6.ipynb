Now we are ready for classification

```python
X=df[features]
Y=df['Species']
```

```python
X
```

```python
Y
```

```python
#As y has str values we convert it into numeric value so that it can be fittted in the model
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()
y=enc.fit_transform(Y)
```

```python
y
```

Thus we can see the encoder had given values - 
>
0 - Iris setosa
>
1 - Iris versicolor
>
2 - Iris virginica

```python
#Creating training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
#Selecting Model
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='gini', splitter='best',random_state=42)
```

```python
#Fitting the data
plt_tree=model.fit(x_train,y_train)
```

```python
#Getting model accuracy
print(f"Accuracy of Decision Tree Classifier is {model.score(x_test,y_test)*100:.2f}%")
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(plt_tree,feature_names=features)
plt.show()
```

```python
#viewing as text
from sklearn.tree import export_text
r=export_text(plt_tree,feature_names=features)
print(r)
```

You can now feed any new/test data to this classifer and it would be able to predict the right class accordingly.

# Predicting with the help of model created

Now let us evaluate the model metrics

```python
from sklearn.metrics import precision_score, recall_score, f1_score
print("Precision Score : ",round(precision_score(y_test,y_pred,average='micro'),2))
print("Recall Score :" , round(recall_score(y_test, y_pred,average='macro'),2) )
print("F1 Score : " ,round(f1_score(y_test, y_pred,average='macro'),2))
```

From the above, we can say that we've a pretty good model for Classification.
